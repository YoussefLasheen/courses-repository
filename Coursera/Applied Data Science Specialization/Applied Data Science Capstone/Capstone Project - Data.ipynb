# Capstone Project - Cafes in Luxor
## Introduction:
In the city of Luxor, if an investor is willing to open a buisness, A Cafe for example , the problem is, the place would you suggest that they open it? The foundation of the issue is that all together for a Cafe to be productive, there must be sufficient customers, and so as to have enough customers, it is not sensible setting up a Cafe in proximity to existing ones.

## Problem
- Finding the ideal location to open a new buisness

## Solution
- Find the location that has a lot of customers that are willing to visit the cafe. 
- And in the same time not being too close to simmilar buisnesses.

## Data Description:
The data used to solve this problem is geolocation data collected from FourSquare. 
Data is a single dataframe, containing at least a location of the café. Explanation of the location data is a standard tuple (lat, lng).
