## Dataset Overview

The data consists of information regarding 183,412 fordgo bikes trips covering the greater San Francisco Bay area in Febuary 2019, including trip start_date, trip duration, and other user realted attributes. The dataset can be found [here](https://video.udacity-data.com/topher/2020/October/5f91cf38_201902-fordgobike-tripdata/201902-fordgobike-tripdata.csv), with feature documentation available [here](https://docs.google.com/document/d/e/2PACX-1vQmkX4iOT6Rcrin42vslquX2_wQCjIa_hbwD0xmxrERPSOJYDtpNc_3wwK_p9_KpOsfA6QVyEHdxxq7/pub).

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('201902-fordgobike-tripdata.csv')
```

## Gender distribution for users


By looking at the chart below, we can see that most fequent users are males. 

```python
plt.figure(figsize=[ 11.69, 8.27])
plt.bar(x = df.member_gender.value_counts().keys(), height = df.member_gender.value_counts() )
plt.title('Gender distribution')
plt.ylabel('No. of Users')

plt.show();
```

## user-type distribution for users

By looking at the chart below, we can see that most users are subsribers other than Customers. 

```python
plt.figure(figsize=[ 11.69, 8.27])

sb.countplot(data = df, x = 'user_type').set_title('user-type distribution for users');
```

## Duration in seconds relation to user type
Trip Duration vs. User Type
By looking at the chart below, we can see that Customers have longer trips opn average compared to subscribers. Most of the subscribers take less than 500 sec trips

```python
df = df.query('duration_sec <= 3176')
plt.figure(figsize=[ 11.69, 8.27])
sb.violinplot(data=df, x='user_type', y='duration_sec')
plt.xlabel('User Type')
plt.ylabel('Trip Duration(sec)')
plt.title('Duration in seconds relation to user type')
plt.show()
```
